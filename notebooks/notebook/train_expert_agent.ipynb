```python
import sys
sys.path.append("../")
import os
```

```python
from simulation.farm_env import FarmEnv
from stable_baselines3 import PPO
from stable_baselines3.common.callbacks import CheckpointCallback
```

```python
TRAINING_STEPS = 500000
MODEL_NAME = "ppo_real_weather_final.zip"
```

```python
models_dir = "../models"
logs_dir = "../logs"
os.makedirs(models_dir, exist_ok=True)
os.makedirs(logs_dir, exist_ok=True)
```

```python
print("Creating evironment with real weather data...")
env = FarmEnv()
env.reset()
print("Environment ready.")
```

```python
model = PPO("MlpPolicy", env, verbose=1, tensorboard_log=logs_dir)

checkpoint_callback = CheckpointCallback(
    save_freq=50000,
    save_path=logs_dir,
    name_prefix="real_weather_model_checkpoint"
)

print(f"Starting training for {TRAINING_STEPS} timestamps...")
model.learn(
    total_timesteps=TRAINING_STEPS,
    callback=checkpoint_callback,
    tb_log_name="PPO_RealWeather"
)
```

```python
final_model_path = os.path.join(models_dir, MODEL_NAME)
model.save(final_model_path)

print(f"\n Training Completed")
print(f"Expert Model saved at: {final_model_path}")
```
